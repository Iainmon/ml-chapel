```python
import struct
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

def load_tensors_from_file(path, num_tensors=1):
    tensors = []
    with open(path, mode='rb') as file:
        # file_content = file.read()
        for i in range(num_tensors):
            rank, = struct.unpack('q', file.read(8))
            shape = []
            for i in range(rank):
                dim, = struct.unpack('q', file.read(8))
                shape.append(dim)
            
            a = np.zeros(shape)
            for i in range(np.prod(shape)):
                r, = struct.unpack('d', file.read(8))
                a.flat[i] = r
            tensors.append(a)
    return tensors
```

```python
def run():
    layers = load_tensors_from_file('filters/filter.bin',num_tensors=2)

    for filters in layers:
        num_filters = filters.shape[0]
        for i in range(num_filters):
            num_kernels = filters.shape[-1]
            for k in range(num_kernels):
                plt.subplot(num_filters, num_kernels, i*num_kernels+k+1)
                plt.imshow(filters[i,:,:,k],cmap=plt.get_cmap('gray'))
                plt.axis('off')
        plt.show()
```
